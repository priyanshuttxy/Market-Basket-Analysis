```python
l=[]
for i in range(0,len(data)):
    l1=[]
    for j in range(0,20):
        l1.append(str(data.iloc[i,j]))
    l.append(l1)
```

```python
len(l)
```

A rule with a lift of 3 implies that knowing one item occurs makes it three times more likely that the associated item will also occur. For example, in a retail context, if the rule is {Bread} -> {Butter} with a lift of 3, it means customers buying bread are three times more likely to buy butter compared to random customers.


Min length refer to the number of items in which the association rules need to apply

Min_support & min_confidence gives the threshold value for the association to be concidered

```python
as_model=apriori(l,min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)
list_as=list(as_model)
```

```python
list_as
```

```python
df=pd.DataFrame()
prd1=[]
prd2=[]
supp=[]
con=[]
lift=[]
for item in list_as:
    lst=[]
    for x in item[0]:
        lst.append(x)
    prd1.append(lst[0])
    prd2.append(lst[1])
    supp.append(item[1])
    con.append(item[2][0][2])
    lift.append(item[2][0][3])
```

```python
df["Procuct1"]=prd1
df["Procuct2"]=prd2
df["Support"]=supp
df["Confidence"]=con
df["Lift"]=lift
```

```python
df.sort_values(by=["Lift"], ascending=False)
```

```python
df.drop(df[(df.Procuct1 == "nan") | (df.Procuct2 == "nan")].index, inplace=True)
```

```python
df.to_csv("Result_MBA.csv")
```
